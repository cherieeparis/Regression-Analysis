## (a) Run an OLS regression, including at least one independent variable and a time variable (as dummies).  Explain how you think your independent variable relates to your dependent variable.  Interpret your results.  Did you find what you expected to find? 

```python
#runs an OLS regression, including at least one independent variable and a time variable (as dummies) 

lm1 = smf.ols('educ ~ paeduc + year', data=sub).fit()
lm1.summary()
```

### for this regression model, I chose paedu as independent variable, year as dummies variable educ as dependent variable that I want to see how father's education will affect their kids' education attainment. I assume, people will have higher education attainment with year passes, also their father's degree will positive affect them to gain higher education. In order to verify my assumption, we can look at the output from the linear regression model. 
### R-square = 0.258, it presents that 25.8% variance of educ can be explained by paedu and year. Constant the year, the father gain one addtional unit of education, their kid's education will raise up by 0.3297 unit, at a statistcally significant level with p-value =0. Constant the paeduc, the respondent's education goes up with year passes by 0.0318 unit, at a statisitcally significant. Therefore, we can say paeduc and year have positive correlation with educ and it's less likely to happen it by chance with this low se and high t-stats value, and we can reject null hypothesis.

# Then run a random effects model equivalent to your fixed effects model in step (b).  Interpret the results.

```python
from linearmodels.panel import RandomEffects

# run the random effects model equivalent from lm2
lm4 = RandomEffects.from_formula('educ ~ paeduc + C(year)', data=sub).fit()
lm4.summary
```

### In this model, R-square = 0.2599, which is lower than lm3(0.298), it tells that 25.99% of variance in educ can be explained by paeduc and year. parameter paeduc = 0.3284, it is same as the coefficient of paeduc in lm2 from previous in fixed effect model. It has a very low p-value (0.0000), suggesting a statistically significant positive relationship between father education and the education level of the individual.
### For years, all of these coefficients are highly statistically significant (p-values are 0.0000). For year 1975, the average education level is estimated to be 0.1963 units higher than the reference year(year 1972), constant the paeduc. It suggest an increase in education level for the year 1975 relative to the refernce year and this observed change is unlikely to be due to random chance. After year 1975, the parameter generally presents a trend upward state that with year goes up, the individual education level will raise faster compared with the reference year 1972. 
### R-square (between) = 0.7331, it tells that the proportion of total variability in educ attributed to differences between paedu that grouped by year. R-square(within) = 0,2533, it tells that the proportion of total variability in educ attributed to differences within paeduc over time. It indicate that a substantial portion of the total variability in educ is attributed to difference between years and within years over time. 
### Overall, the random effects model suggests that both father's education (paeduc) and the year of observation (C(year)) have statistically significant effects on the education level (educ). 

## Run a Hausman test to compare your fixed effects and your random effects models.  What do you conclude? 

```python
from linearmodels.panel import PanelOLS, RandomEffects

y = sub['educ']
x = sub[['paeduc', 'year']]

# Fixed Effects Model
lm2 = PanelOLS(y, x, entity_effects=True, time_effects=True).fit()

# Random Effects Model
lm4 = RandomEffects(y, x).fit()

# Wald test for exogeneity in Random Effects Model
wald_test = lm4.wald_test(formula='paeduc = 0, year = 0')

# Print Wald test results
print(wald_test)
```

### in this case, the p-value from the test is very small, it suggests that the fixed effects model is preferred.  Also, with the pvalue = 0, we can reject the null hypothesis, indicating that the coefficients of the specified variables are not jointly equal to zero. 
### compared with lm2(fixed effect model) and lm4(random effect model), the R-square is higher on lm2 with the constant value on paedu of coeff. Also, lm2 has better clear on dummy variable year that it tells us specific time period we could focus and observe the effect between year and educ, while lm4 didn't work well on it as it shows entire time period are statistically significant.

### In the output, R-square =0.306, which means 30.6% of variation of child' education is explained by their  parent education from both side. Constant the father education, there is positive affect between educ and maeduc, on average, with additional one unit of mother education increase, their child education will increase 0.2131 unit, with a statistically significant.  t-stats = 10.714, which means there is low posibility that it have had happen by chance with this small standard error. Therefore, there has difference between educ and maeduc. 
### Constant the mother education, there is positive affect between educ and paeduc. On average, with additional one unit of father education increase, their child education will increase 0.2311 unit, with a statistically significant.  t-stats = 13.072, which means there is low posibility that it have had happen by chance with this small standard error.
